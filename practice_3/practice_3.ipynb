### 2)Написать функцию, которая производит поиск и выводит на экран вакансии с заработной платой больше введенной суммы

```python
def find_by_salary(db, table_name, salary=None, currency_name='руб'):
    for vacancy in db.table_name.find({'$and': [
                                                {'currency_name': currency_name}, 
                                                {'$or':[
                                                        {'min': {'$gte':salary}}, 
                                                        {'max': {'$gte':salary}}
                                                        ]
                                                }
                                                ]
                                      }):
        pprint(vacancy)
        
```

```python
find_by_salary(db, 'vacancies_table', 2000, 'USD')
```

### 3*)Написать функцию, которая будет добавлять в вашу базу данных только новые вакансии с сайта

```python
def fill_only_new_vacancies(db, table_name):
    data = get_data_from_practice_2()
    for vacancy in data:
        db.table_name.update_one({'href': vacancy['href']}, {'$set':vacancy}, upsert=True)
    
```

```python
fill_only_new_vacancies(db, 'vacancies_table')
```
